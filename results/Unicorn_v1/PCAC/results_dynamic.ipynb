# G-PCC

```python
csvdir_list = ['csvfiles/dynamic/gpcc/raht21/soldier_seqs.csv',
                'csvfiles/dynamic/gpcc/raht21/ball_seqs.csv']

file_list = [csvdir_list[i] for i in seqs_idxs]
for i, f in enumerate(file_list):print(i+1, f)
df = mean_file(file_list)
x = np.array(df[xlabel+'_att'])[:-2].round(3)
y = np.array(df[ylabel])[:-2].round(3)

enc_time = np.array(df['Enc_colors processing time (user)'])[:-2].round(3)
dec_time = np.array(df['Dec_colors processing time (user)'])[:-2].round(3)

enc_timeG = np.array(df['Enc_positions processing time (user)'])[:-2].round(3)
dec_timeG = np.array(df['Dec_positions processing time (user)'])[:-2].round(3)

print('x', [i for i in x])
print('y', [i for i in y])

print('enc_time', [i for i in enc_time], enc_time.mean().round(1))
print('dec_time', [i for i in dec_time], dec_time.mean().round(1))

print('enc_timeG', [i for i in enc_timeG], enc_timeG.mean().round(1))
print('dec_timeG', [i for i in dec_timeG], dec_timeG.mean().round(1))

gpcc = [x, y]
```

```python
linestyle='solid'
metric=ylabel
markersize=14

fig, ax = plt.subplots(figsize=(7.3*2,5*2))
prefix = '  '


plt.plot(ges[0][:], ges[1][:], 
         label='G-PCC (Dynamic)', 
         color='black', marker='>', linestyle='--', linewidth=3+0.5, markersize=markersize+1, markeredgewidth=2)


plt.plot(gpcc[0][:], gpcc[1][:], 
         label='G-PCC (Static)', 
         color='black', marker='o', linestyle=linestyle, linewidth=3, markersize=markersize-1)


plt.plot(inter[0][:], inter[1][:], 
         label='Unicorn (Dynamic)', 
         color='#D11141', marker='*', linestyle='--', linewidth=3+0.5, markersize=markersize+6, markeredgewidth=2)

plt.plot(intra[0][:], intra[1][:], 
         label='Unicorn (Static)', 
         color='red', marker='D', linestyle=linestyle, linewidth=3, markersize=markersize)


ax.locator_params(axis='x', nbins=12)
ax.locator_params(axis='y', nbins=12)
plt.tick_params(labelsize=32)
plt.xlabel(xlabel, fontdict={'size':40})
plt.ylabel(metric, fontdict={'size':40})

plt.title(title, fontdict={'size':46})
plt.grid(ls='-.')
plt.legend(loc='lower right', ncol=1, prop={'size':36})

fig.tight_layout()
os.makedirs('figs/dynamic/', exist_ok=True)
fig.savefig( os.path.join('figs/dynamic/', title+'_'+metric+'.pdf'))
```
