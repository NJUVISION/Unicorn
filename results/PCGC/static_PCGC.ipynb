# Ours

```python
if any([i in seqs_idxs for i in [0,1,2,3,4,5,6,7,13]]): 
    # solide PCs
    length = 7
elif any([i in seqs_idxs for i in [8,9,10,11,12]]): 
    # dense/sparse
    length = 9
else:
    length = None

csvdir_list = [
'csvfiles/ours/8ivfb/0_longdress_vox10_1300.csv',
'csvfiles/ours/8ivfb/1_loot_vox10_1200.csv',
'csvfiles/ours/8ivfb/2_redandblack_vox10_1550.csv',
'csvfiles/ours/8ivfb/3_soldier_vox10_0690.csv',

'csvfiles/ours/owlii/0_basketball_player_vox11_00000200.csv',
'csvfiles/ours/owlii/1_dancer_vox11_00000001.csv',
'csvfiles/ours/owlii/2_exercise_vox11_00000001.csv',
'csvfiles/ours/owlii/3_model_vox11_00000001.csv',

'csvfiles/ours/mpeg_dense/0_Facade_00009_vox12.csv',
'csvfiles/ours/mpeg_dense/1_House_without_roof_00057_vox12.csv',

'csvfiles/ours/mpeg_sparse/0_Arco_Valentino_Dense_vox12.csv',
'csvfiles/ours/mpeg_sparse/1_Shiva_00035_vox12.csv',
'csvfiles/ours/mpeg_sparse/2_Staue_Klimt_vox12.csv',

'csvfiles/ours/scan2cm.csv',

'csvfiles/ours/ford1mm.csv',
'csvfiles/ours/kitti1mm.csv']

file_list = [csvdir_list[i] for i in seqs_idxs]
for i, f in enumerate(file_list):print(i+1, f)
df = mean_file(file_list)

if title in ['Ford_q1mm', 'KITTI_q1mm', 'LiDAR_q1mm (Ford,KITTI)'] and ylabel not in ['enc_time', 'dec_time']:
    curr_ylabel = 'offset_'+ylabel
else:
    curr_ylabel = ylabel

x = np.array(df[xlabel])[:length].round(3)
y = np.array(df[curr_ylabel])[:length].round(3)

print('x', [i for i in x])
print('y', [i for i in y])

ours = [x, y]
```

# RD Curve

```python
if ylabel=='mseF,PSNR (p2point)': ylabelname = 'D1-PSNR'
if ylabel=='mseF,PSNR (p2plane)': ylabelname = 'D2-PSNR'

if ylabel=='enc_time': ylabelname = 'Encoding Time (s)'
if ylabel=='dec_time': ylabelname = 'Decoding Time (s)'

linestyle='solid'

# title = title + '  '+ylabel
fig, ax = plt.subplots(figsize=(7.3*2,5*2))
prefix = ' '

plt.plot(ours[0][1:], ours[1][1:], 
         label='Unicorn (Static)', 
         color='#F75488', marker='o', linestyle=linestyle, linewidth=3, markersize=12)

plt.plot(gpcc[0][1:], gpcc[1][1:], 
         label='G-PCC (Static)', 
         color='#6BB5EE', marker='o', linestyle=linestyle, linewidth=2, markersize=12)


ax.locator_params(axis='x', nbins=12)
ax.locator_params(axis='y', nbins=12)
plt.tick_params(labelsize=32)
plt.xlabel(xlabel, fontdict={'size':40})
plt.ylabel(ylabelname, fontdict={'size':40})

plt.title(title, fontdict={'size':46})
plt.grid(ls='-.')
plt.legend(loc='lower right', ncol=1, prop={'size':30})


fig.tight_layout()

if ylabel in ['mseF,PSNR (p2point)', 'mseF,PSNR (p2plane)']: 
    os.makedirs('figs/', exist_ok=True)
    fig.savefig( os.path.join('figs/', title+'_'+ylabelname+'.pdf'))
    
```
